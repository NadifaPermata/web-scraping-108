```python
import pandas as pd
from bs4 import BeautifulSoup

import chromedriver_autoinstaller
from selenium import webdriver
```

```python
chromedriver_autoinstaller.install()
browser = webdriver.Chrome()
```

```python
browser.get("https://www.gutenberg.org/")
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
soup
```

```python
data_judul_cover=[]
data_penulis=[]
data_judul=[]

judul_cover = soup.find_all('div', class_='cover_title')
for cover in judul_cover:
    data_judul_cover.append(cover.h5.text)

penulis = soup.find_all('div', class_='cover_img')
for authors in penulis:
    data_penulis.append(authors.img['authors'])

judul = soup.find_all('div', class_='cover_img')
for title in judul:
    data_judul.append(title.img['title'])
```

```python
tabel = {
    'cover' : data_judul_cover,
    'penulis' : data_penulis,
    'judul' : data_judul
}
```

```python
pd.DataFrame(tabel)
```

```python
tabel = {
    'cover' : data_judul_cover,
    'penulis' : data_penulis,
    'judul' : data_judul
}
pd.DataFrame(tabel).to_excel('kutip.xlsx')
```
